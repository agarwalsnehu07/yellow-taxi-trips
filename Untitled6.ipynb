```python
import pandas as pd
df = pd.read_csv("yellow_tripdata_2023-01.csv",
                    parse_dates=["tpep_pickup_datetime", "tpep_dropoff_datetime"],
                    low_memory=False)
df.dropna(inplace=True)
df = df[(df['trip_distance'] > 0) & (df['trip_distance'] < 50)]
df = df[(df['fare_amount'] > 0) & (df['fare_amount'] < 500)]
```

```python
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python

df['speed_mph'] = df['trip_distance'] / (df['trip_duration'] / 60)
```

```python
df['time_of_day'] = pd.cut(
    df['tpep_pickup_datetime'].dt.hour,
    bins=[0, 6, 12, 18, 24],
    labels=['Night', 'Morning', 'Afternoon', 'Evening'],
    right=False
)
```

```python
df['is_weekend'] = df['tpep_pickup_datetime'].dt.weekday >= 5
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['pickup_date'] = df['tpep_pickup_datetime'].dt.date
daily_revenue = df.groupby('pickup_date')['total_amount'].sum()

plt.figure(figsize=(12, 6))
sns.lineplot(x=daily_revenue.index, y=daily_revenue.values, label='Daily Revenue')
plt.plot(daily_revenue.index, daily_revenue.rolling(7).mean(), label='7-day Moving Average', linestyle='dashed')

plt.xlabel("Date")
plt.ylabel("Total Revenue ($)")
plt.title("Daily Taxi Revenue - January 2023")
plt.legend()
plt.show()
```

```python
max_date = daily_revenue.idxmax()
min_date = daily_revenue.idxmin()

plt.figure(figsize=(12, 6))
sns.lineplot(x=daily_revenue.index, y=daily_revenue.values, label="Daily Revenue")
plt.plot(daily_revenue.index, daily_revenue.rolling(7).mean(), label="7-day Moving Average", linestyle="dashed")

plt.annotate(f'Peak: {max_date}', xy=(max_date, daily_revenue[max_date]), xytext=(max_date, daily_revenue[max_date] + 20000),
             arrowprops=dict(facecolor='green', arrowstyle='->'))
plt.annotate(f'Trough: {min_date}', xy=(min_date, daily_revenue[min_date]), xytext=(min_date, daily_revenue[min_date] - 20000),
             arrowprops=dict(facecolor='red', arrowstyle='->'))

plt.xlabel("Date")
plt.ylabel("Total Revenue ($)")
plt.title("Daily Taxi Revenue with Peaks and Troughs - January 2023")
plt.legend()
plt.show()
```

```python
df.groupby("time_of_day")["trip_duration"].mean().plot(kind="bar", title="Avg Trip Duration by Time of Day")
```

```python
df['day_of_week'] = df['tpep_pickup_datetime'].dt.day_name()
df.groupby('day_of_week')['total_amount'].sum().plot(kind='bar', title="Revenue by Day of the Week")
```

```python
sns.histplot(df['speed_mph'], bins=30, kde=True)
plt.title("Distribution of Taxi Speeds")
```
